```python
from elasticsearch import Elasticsearch
```

```python
es = Elasticsearch(
    hosts=[{'host': 'localhost', 'port': 9200, 'scheme': 'http'}]
)
```

```python

info = es.info()
print(info)
```

```python
import pandas as pd
df=pd.read_csv('Data/myntra_products_catalog.csv')
df
```

```python
df.describe()
```

```python
df.isna().value_counts()
```

```python
df.fillna("None", inplace=True)
```

```python
df.isna().value_counts()
```

```python
from sentence_transformers import SentenceTransformer

# Load https://huggingface.co/sentence-transformers/all-mpnet-base-v2
model = SentenceTransformer("all-mpnet-base-v2")
```

```python
df["DescriptionVector"]=df["Description"].apply(lambda x: model.encode(x))
```

```python
df.head()
```

```python
es.ping()
```

```python
##create new index in elastic search
from index.indexmapping import indexmapping
es.indices.create(index="all_products", mappings=indexmapping)
```

```python
## ingest the data into index
df
```

```python
 record_list=df.to_dict("records")
```

```python
record_list[0]
```

```python
for record in record_list:
    try:
        es.index(index="all_products",document=record, id= record["ProductID"])
    except Exception as e:
            print(e)
```

```python
es.count(index="all_products")
```

```python
input="traditional kurties"
vector_of_input=model.encode(input)
query={
    "field":"DescriptionVector",
    "query_vector":vector_of_input,
    "k":2,
    "num_candidates":500,
}

res=es.knn_search(index="all_products",knn=query,source=["ProductName","Description"])
res["hits"]["hits"]
```
